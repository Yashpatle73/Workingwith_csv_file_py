# Working with CSV Files 

```python
import csv
import matplotlib.pyplot as plt
runs=[]
areas=[]
with open('cricketcsv.csv','r') as fr:
    data=csv.reader(fr)
    next(data)
    for row in data:
        areas.append(row[0])
        runs.append(int(row[1]))
c=['c','y','m','r','b']
e=[0,0.1,0,0,0]
plt.pie(runs,labels=areas,autopct='%.1f%%',colors=c,explode=e,shadow=True)
plt.show()
```
